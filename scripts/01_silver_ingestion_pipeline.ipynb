```python
import urllib.parse

# ==========================================
# 1. SECURITY & AUTHENTICATION
# ==========================================
print("🔒 Fetching credentials from Databricks Secret Scope...")

# Fetch the secrets securely from the vault we built
aws_access_key = dbutils.secrets.get(scope="lakehouse_secrets", key="aws_access_key")
aws_secret_key = dbutils.secrets.get(scope="lakehouse_secrets", key="aws_secret_key")
aws_bucket_name = "upi-payments-lakehouse"

# Safely URL-encode the secret key so Hadoop doesn't misread special characters
encoded_secret_key = urllib.parse.quote(aws_secret_key, safe="")

# Inject the variables to create the secure connection string
s3_path = f"s3a://{aws_access_key}:{encoded_secret_key}@{aws_bucket_name}/raw_upi_data/"

print(f"✅ Secure connection string generated for: {aws_bucket_name}")
```
