```python
import os
from dotenv import load_dotenv

# ==========================================
# 1. LOCAL SECURITY & AUTHENTICATION
# ==========================================
print("🔒 Fetching credentials from local .env vault...")

# Load the environment variables from your .env file
load_dotenv()

# Fetch the secrets safely
aws_access_key = os.getenv("AWS_ACCESS_KEY_ID")
aws_secret_key = os.getenv("AWS_SECRET_ACCESS_KEY")
aws_bucket_name = os.getenv("S3_BUCKET_NAME")

# A safe check to prove it worked without printing the whole password
if aws_access_key and aws_secret_key:
    print(f"✅ Secrets loaded successfully!")
    print(f"✅ Target Bucket: {aws_bucket_name}")
    print(f"🔑 Access Key ID starts with: {aws_access_key[:4]}********")
else:
    print("❌ Error: Could not find keys. Check your .env file!")
```
